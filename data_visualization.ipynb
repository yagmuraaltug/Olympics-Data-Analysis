```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('olympics.csv')
data.head()
```

SCATTER GRAPH

```python
def drawscatter():
    sns.scatterplot(x='Height', y='Weight', data=data) # x, y vectors or keys in data
    plt.title('Relationship Between Height and Weight')
    plt.show()
drawscatter()
```

```python
sns.set_style('darkgrid')
drawscatter()
```

```python
sns.scatterplot(x='Height', y='Weight',hue='Medal' ,data=data) # hue vector or key in data
plt.title('Medals According to Height and Weight')
plt.show()
```

```python
sns.scatterplot(x='Height', y='Weight',style='Medal', hue='Gender' ,data=data)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height and Weight of Sporters')
plt.show()
```

```python
sns.set_style('whitegrid')
sns.scatterplot(x='Height', y='Weight',style='Medal', hue='Gender',size='Age', sizes=(15,200) ,data=data)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height-Weight Changes of Sporters')
plt.show()
```

LINEAR REGRESSION
Plot data and a linear regression model fit.

```python
sns.regplot(x='Height', y='Weight', data=data, marker='+', scatter_kws={'alpha':0.2}) # upword slope
plt.title('Height and Weight Distribution')
plt.show()
```

```python
sns.scatterplot(x='Height', y='Weight', hue='Medal' ,data=data, palette='Set1')
plt.title('Height and Weight Distribution According to Medal')
plt.show()
```

```python
sns.lineplot(x='Height', y='Weight', data=data)
plt.title('Height-Weight')
plt.show()
```

HISTOGRAM

Figure-level interface for drawing distribution plots onto a FacetGrid.

```python
sns.displot(data, x='Weight')
plt.ylabel('Frequency')
plt.title('Histogram of Weight')
plt.show()
```

```python
sns.displot(data, x='Weight', hue='Gender')
plt.ylabel('Frequency')
plt.title('Histogram of Weight')
plt.show()
```

```python
sns.displot(data, x='Weight', col='Gender', multiple='dodge')
plt.show()
```

TWO-DIMENSIONAL HISTOGRAM

```python
sns.displot(data, x='Weight', y='Height',hue='Gender', kind='kde')
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title('Histogram of Weight-Height')
plt.show()
```

KDE PLOT 

Plot univariate or bivariate distributions using kernel density estimation.

```python
sns.kdeplot(data=data, x='Weight', y='Height',hue='Gender', fill=True)
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title('Histogram of Weight-Height')
plt.show()
```

BAR PLOT

```python
sns.barplot(x='Medal', y='Height', hue='Gender',data=data)
plt.title('Height According to Medals')
plt.show()
```

CAT PLOT


```python
sns.catplot(x='Medal', y='Age', hue='Gender',col='Season',kind='bar',height=4,aspect=0.7, data=data)
plt.show()
```

BOX PLOT

```python
sns.boxplot(x='Season', y='Height',hue='Gender', data=data, palette='Set2')
plt.title('Height According to Medals')
plt.show()
```

```python
data_copy = data.loc[:,['Age','Weight','Height']]
sns.boxplot(data=data_copy,orient='h',palette='rocket') #“v” | “h”, optional Orientation of the plot (vertical or horizontal).
plt.show()
```

```python
sns.catplot(x='Season', y='Height', hue='Gender',col='Medal',kind='box',height=4,aspect=0.7, data=data)
#Aspect ratio of each facet, so that aspect * height gives the width of each facet in inches.
plt.show()
```

HEATMAP 

```python
sns.heatmap(data.corr(), annot=False, linewidths=0.5, fmt='.1f') #String formatting code to use when adding annotations.
plt.show()
```

VIOLIN PLOT

```python
sns.violinplot(x='Season',y='Height',hue='Gender',data=data,split=True)
plt.show()
```

```python
sns.catplot(x='Season',y='Height',hue='Gender',col='Medal', data=data, split=True, kind='violin', height=4,aspect=0.7)
plt.show()
```

JOINT PLOT

```python
sns.jointplot(data=data,x='Weight', y='Height',hue='Season',kind='kde')
plt.show()
```

```python
g = sns.JointGrid(data=data,x='Weight', y='Height') #Grid for drawing a bivariate plot with marginal univariate plots.
g.plot_joint(sns.histplot)  #If need to pass different keyword arguments to each function,  have to invoke JointGrid.plot_joint() and JointGrid.plot_marginals():
g.plot_marginals(sns.boxplot)
plt.show()
```

PAIR GRAPH

```python
sns.pairplot(data)
plt.show()
```

COUNT GRAPH

```python
sns.countplot(x='City',data=data)
plt.xticks(rotation = 90)
plt.show()
```
